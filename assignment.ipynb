```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
drugs = pd.read_csv('drug200.csv')
from sklearn.model_selection import train_test_split
xTrain , xTest, yTrain , yTest = train_test_split(drugs.iloc[:,:5],drugs.iloc[:,-1], test_size=0.2)
```

```python
xTest=pd.get_dummies(xTest, columns=['Sex'],drop_first=True)
xTest['Sex_M']=xTest['Sex_M'].astype(int)
xTrain=pd.get_dummies(xTrain, columns=['Sex'],drop_first=True)
xTrain['Sex_M']=xTrain['Sex_M'].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(yTrain)
yTrain=le.transform(yTrain)
yTest=le.fit_transform(yTest)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['LOW', 'NORMAL', 'HIGH']])
xTrain['BP_NUM'] = oe.fit_transform(xTrain[['BP']])
xTrain['Cholesterol_NUM'] = oe.fit_transform(xTrain[['Cholesterol']])
xTrain=xTrain.drop('BP', axis=1)
xTrain=xTrain.drop('Cholesterol', axis=1)

xTrain
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=[['LOW', 'NORMAL', 'HIGH']])
xTest['BP_NUM'] = oe.fit_transform(xTest[['BP']])
xTest['Cholesterol_NUM'] = oe.fit_transform(xTest[['Cholesterol']])
xTest=xTest.drop('BP', axis=1)
xTest=xTest.drop('Cholesterol', axis=1)

xTest
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(xTrain,yTrain)
```

```python
reg.score(xTest,yTest)
```

```python
Logistic = linear_model.LogisticRegression(max_iter=10000)
```

```python
Logistic.fit(xTrain, yTrain)
Logistic.score(xTest, yTest)
```

```python
from sklearn.naive_bayes import GaussianNB
svm=GaussianNB()
svm.fit(xTrain, yTrain)
svm.score(xTest, yTest)
```
